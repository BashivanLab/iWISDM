```python
import pandas as pd
import os
import shutil
from natsort import natsorted, ns

DIR = 'data/subshapenet/subshapenet/val_new_obj_obj_mod'
newDIR = 'data/subshapenet/subshapenet/val'


if os.path.exists(newDIR):
    shutil.rmtree(newDIR)
os.makedirs(newDIR)
```

```python
df = pd.read_pickle('data/subshapenet/subshapenet/val_new_obj_meta_data.pkl').set_index('ref').reset_index()
df
```

```python
id = 0
for obj_mod in natsorted(os.listdir(DIR)):
    fp = os.path.join(DIR, obj_mod)
    print(fp)
    if not obj_mod.isnumeric():
        continue
    for image in natsorted(os.listdir(fp)):
        if image[-5:] != '.jpeg':
            continue
        imf_fp = os.path.join(fp,image)
        ref = int(df.iloc[id]['ref'])
        new_img_fp = newDIR + '/' + str(ref)
        os.mkdir(new_img_fp)
        shutil.copyfile(imf_fp, new_img_fp + '/image.png')
        id += 1
```
